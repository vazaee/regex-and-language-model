```python
def remove(texts, regex):
    if type(texts) == str:
        return regex.sub("", texts)
    else:
        return [regex.sub("", text) for text in texts]
```

```python
def change_code(texts, regex):
    if type(texts) == str:
        return regex.sub("CODE", texts)
    else:
        return [regex.sub("CODE", text) for text in texts]
```

```python
def to_lower(texts):
    if type(texts) == str:
        return texts.lower()
    else:
        return [text.lower() for text in texts]
```

```python
def change_by_space(texts, regex):
    if type(texts) == str:
        return regex.sub(" ", texts)
    else:
        return [regex.sub(" ", text) for text in texts]
```

```python
def train_mle_model(texts_list):
    all_questions = ' '.join(texts_list)
    all_words = WhitespaceTokenizer().tokenize(all_questions)
    bigrams, vocabulary = padded_everygram_pipeline(2, all_words)
    model = MLE(2)
    model.fit(bigrams, vocabulary)
    return model
```

```python
def calculate_perplexity(model, text):
    perplexity = 0
    words = WhitespaceTokenizer().tokenize(text)
    words_fakechar = [list(pad_both_ends(word, n=2)) for word in words]
    words_bigrams = [list(bigrams(word)) for word in words_fakechar]

    for word in words_bigrams:
        perplexity += model.perplexity(word)

    return perplexity
```

```python
from nltk.lm import Laplace
```

```python
def train_laplace_model(texts_list):
    all_questions = ' '.join(texts_list)
    all_words = WhitespaceTokenizer().tokenize(all_questions)
    bigrams, vocabulary = padded_everygram_pipeline(2, all_words)
    model = Laplace(2)
    model.fit(bigrams, vocabulary)
    return model
```

```python
eng_model_laplace = train_laplace_model(eng_train)
span_model_laplace = train_laplace_model(span_train)
port_model_laplace = train_laplace_model(port_train)
print(calculate_perplexity(eng_model_laplace, port_test.iloc[1]))
print(calculate_perplexity(span_model_laplace, port_test.iloc[1]))
print(calculate_perplexity(port_model_laplace, port_test.iloc[1]))
```

```python
def set_language(texts_list):
    language = []
    for text in texts_list:
        port = calculate_perplexity(port_model_laplace, text)
        eng = calculate_perplexity(eng_model_laplace, text)
        span = calculate_perplexity(span_model_laplace, text)

        if eng >= port <= span:
            language.append('portuguese')
        elif port > eng < span:
            language.append('english')
        else:
            language.append('spanish')
    
    return language
```

```python
portuguese_results = set_language(port_test)
portuguese_rate = portuguese_results.count("portuguese")/len(portuguese_results)
print(portuguese_rate)
```

```python
english_results = set_language(eng_test)
english_rate = english_results.count("english")/len(english_results)
print(english_rate)
```

```python
spanish_results = set_language(span_test)
spanish_rate = spanish_results.count("spanish")/len(spanish_results)
print(spanish_rate)
```
